```python
import matplotlib.pyplot as plt
import numpy as np
```

## linear plot

```python
x = [1,2,3,4,5]
y = [25,10,45,27,16]

plt.plot(x,y, color = 'r')

plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')

```

## bar plot

```python
x = ['py','c','cpp','java']
y = [90,60,65,74]

plt.xlabel('language', fontsize = 10)
plt.ylabel('scale',fontsize = 10)
plt.title('barchart',fontsize = 10)
plt.bar(x,y, color='r', width = 0.6, edgecolor='b', linewidth=2, linestyle=':',label = "popularity")
plt.legend()
plt.show()
```

```python
company  = ['google','amzn','msft','fb']
revenue = [90,136,89,30]
profit = [60,30,66,28]
```

```python
 #Plots (like bar charts) need numerical x-positions, not strings.
# So instead of directly plotting with company names, we assign each company an index (0, 1, 2, ...).Then, 
# we can later replace the tick labels with the company names.

xpos = np.arange(len(company))    
xpos
```

```python
plt.xticks(xpos, company)

plt.bar(xpos-0.2, revenue, label = 'revenue', width= 0.4 )
plt.bar(xpos+0.2, profit, label ='profit', width=0.4)

plt.savefig('bar.png')
plt.legend()
plt.show()
```

```python
plt.yticks(xpos, company)

plt.barh(xpos-0.2, revenue, label = 'revenue')
plt.barh(xpos+0.6, profit, label ='profit')

plt.savefig('horizbar.png')
plt.legend()
plt.show()
```

## scatter plot

```python
day= [1,2,3,4,5,6,7]
temp=[14,95,31,55,52,16,78]

plt.title('scatter plot')
plt.xlabel('Day')
plt.ylabel('Temperature')

plt.scatter(day,temp, s = 190,  marker="^", edgecolor='y',)

plt.savefig('scatterplt.png')
```

```python
blood_sugar_man = [ 97,  86, 151, 122, 117, 113,  86, 131, 162, 152, 136, 108, 151,
        76,  72, 166, 133, 153, 168, 104, 114,  89, 105, 160,  94, 151,
       137,  77, 133, 109]

blood_sugar_woman = [128, 144, 144,  77, 101,  79, 120, 126,  94, 119, 141, 127,  90,
       100, 127, 149,  79, 143, 124, 148, 132, 129, 125,  83,  96, 113,
       147,  94, 139,  72]
```

```python
plt.title('blood sugar')


# for a single sample
plt.hist(blood_sugar_man, rwidth= 0.90, color='g', bins=5, cumulative= -1,label='sugar level')

plt.legend()
```

```python
plt.title('blood sugar')
plt.xlabel('sugar range')
plt.ylabel('total no. of patient')

# for double sample
plt.hist([blood_sugar_man, blood_sugar_woman], rwidth=0.90, color=['g','y'], cumulative= -1,label=['man sugar label','woman sugar label'])

plt.savefig('histplot.png',bbox_inches="tight")
plt.legend()
```

## pie plot

```python
expense =[40,100,56,10,88]
ex_label = ['food','travel','car','bills','others']

plt.pie(expense, labels = ex_label,  autopct="%0.1f%%",radius=0.8, shadow= True, explode=[0.2,0,0,0,0],
       startangle=360)

plt.title('pie chart')

plt.savefig('mypie.png')
plt.legend()
plt.show()
```
