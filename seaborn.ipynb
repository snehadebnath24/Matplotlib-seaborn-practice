```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
## loading dataset of seaborn using pandas
penguin = sns.load_dataset('penguins')
penguin
```

## line plot

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,5,1,8,2]

sns.lineplot(x=var, y=var_1)
plt.show()
```

```python
penguin.head()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data= penguin, hue= 'sex',
             size=20, style='sex')

plt.grid()
plt.title('penguin line plot')
plt.show()
```

```python
sns.lineplot(x='body_mass_g',y='flipper_length_mm',data= penguin,hue='sex',palette=['g','b'])
sns.set_style('whitegrid')
```

## Scatter plot

```python
penguin.head()
```

```python
sns.scatterplot(x='bill_depth_mm',y='body_mass_g',data= penguin,hue= 'island')
sns.set_style('ticks')
```

## strip plot

```python
sns.stripplot(x='species',y='body_mass_g',data= penguin,hue= 'island', dodge=True)
```

## swarm plot

```python
sns.swarmplot(x='species',y='body_mass_g',data= penguin,hue= 'island')
plt.show()
```

## histogram plot

```python
penguin.head(5)
```

```python
sns.histplot(x='body_mass_g', data=penguin, hue='sex',multiple='stack',bins=10)
```

```python
sns.histplot(x='body_mass_g', data=penguin, hue='species',kde= True,multiple= 'dodge')
```

## regression plot

```python
penguin.head()
```

```python
sns.regplot(x='body_mass_g', y='bill_length_mm',data= penguin, color='g', scatter_kws={'alpha':0.6})
```

## joint plot

```python
sns.jointplot(x='body_mass_g',y='flipper_length_mm',data= penguin)
```

```python
sns.jointplot(x='body_mass_g',y='flipper_length_mm',data= penguin, hue='sex')
```

```python
sns.jointplot(x='body_mass_g',y='flipper_length_mm',data= penguin,hue='sex',kind='kde')
```

```python
sns.jointplot(x='body_mass_g',y='flipper_length_mm',data= penguin,kind='reg')
```

## bar plot

```python
penguin.head()
```

```python
sns.barplot(x='species',y='body_mass_g',data= penguin, hue='sex',palette=['skyblue','pink'] ,
           estimator= np.mean)
plt.title('bar plot using seaborn')
plt.show()
```

## count plot

```python
sns.countplot(x='species',data= penguin, hue='island', palette='Set3')
plt.show()
```

## box plot

```python
penguin.head()
```

```python
sns.boxplot(x='island',y='body_mass_g',hue='sex',data= penguin, palette='plasma')
```

```python
plt.figure(figsize=(12, 6)) # adjusting the figure size

sns.boxplot(x='species',y='flipper_length_mm',data= penguin, hue='sex')
plt.show()
```

## violin plot

```python
sns.violinplot(x='species',y='body_mass_g', data= penguin,hue='sex')
```

```python
# combination of swarm and violin plot

sns.violinplot(x='species',y='body_mass_g',data= penguin,color='yellow')
sns.swarmplot(x='species',y='body_mass_g', data= penguin,color='black',marker='*',size=4)

plt.show()
```

## kde plot

```python
penguin.tail()
```

```python
sns.kdeplot(x='body_mass_g', data= penguin, hue='species', fill=True, palette='pastel')
```

## heat map 

```python
penguin.head()
```

```python
# either drop the non-numerical column from the datset like this

x=penguin.drop(columns=['species','island','sex']).head()
x
```

```python
sns.heatmap(x , cmap='Blues', annot=True,linewidth=1, linecolor='black')
```

```python
# or make a list of only numerical columns and make a variable and co relate
# do what is easy going
columns=['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']

penguin[columns].corr()
```

```python
sns.heatmap(data = penguin[columns].corr(), cmap="YlGnBu",annot=True, vmin=-0.2, linewidth=2)
```
